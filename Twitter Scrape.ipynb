```python
#setup the API with your keys
#also, format the return in JSON
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())

#Callingthe search method, with several parameter
#q: querying keyword
#count: limits the numbers of return
result = api.search(q='halloween', count=100)

#grab the tweets from the search result
tweets = result['statuses']

# insert these tweets into the 'tweets' collection
collection.insert_many(tweets)


#now, let's query the tweets we just stored in the collection
#collection.find() will return all records in the collection
records = collection.find()

for record in records:
	#get user and tweet text (note: text must be converted to ascii to display in the console)
	user = record['user']['screen_name']
	tweet = record['text'].encode('ascii','ignore')
	time = record['created_at'].encode('ascii','ignore')
	location = record['user']['location'].encode('ascii','ignore')
	print ('User:', user)
	print ('Tweet:', tweet)
	print ('Time:', time)
	print ('Location:', location)
	print ('\n')
```
